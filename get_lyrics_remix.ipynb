```python
artists = list(df['Artist'])
titles = list(df['Title'])
countries = list(df['Country'])
total_streams = list(df['Total'])
```

```python
artist_titles = [i for i in list(zip(artists, titles))]
```

```python
temp_artist_titles = artist_titles[:10]
```

```python
current_time = datetime.datetime.now()
print(f"Current Time: {current_time}")
artist_title_dict = {}

with open('file.txt', 'w') as file:

    for index in range(starting_index, ending_index):
        artist = artist_titles[index][0]
        title = artist_titles[index][1]
        try:
            lyrics = get_lyrics(artist, title)
        except AttributeError:
            continue
        artist_title_dict[index] = {
            "artist_title": f"{artist} - {title}",
            "lyrics": lyrics,
            "country": countries[index],
            "total_streams": total_streams[index]
        }
        file.write(json.dumps(artist_title_dict[index]))
    end_time = datetime.datetime.now()
    print(f'End Time: {end_time}')

    time_delta = end_time - current_time
    print(f'Fetching Lyrics Process took {time_delta.total_seconds()} seconds.')
```
